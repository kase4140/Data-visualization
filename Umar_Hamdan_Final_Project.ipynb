## Section 2. Data Set Description, Statistics, And Processing

The dataset is in a CSV (Coma Seperated Values) format. We needed to modify the data to fit our demands, By transposing the data and deleting unnecessary information like the title and french translation of the data set, we were able to have a cleaner data set that is easier to work with.

The columns we have are: Year, Married, Common-Law, Separated/Divorced, Single, Other, Total. Each row represent a Year and the number of officers that belong on each marital status category.



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dti

df = pd.read_csv('20190326-marital-status-for-officers.csv')

df.head()
```

## Section 3. Visualization

For the first visualization, We plotted the number of married military officer for each year, we choose a line graph because considering we were plotting progression over time, it felt the most appriopriate for our graphs. Between 2000 and 2003 there was a large influx of married military officers joining the military, after that point it stayed steady, however the largest number of married military officers were 2013 with 8775 married officers.

```python
df2 = df.groupby("Year").mean()
df3 = df2['Married']
years = [1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
scale = [6000,7000,8000,9000]
fig=plt.figure(figsize=(13, 10))
ax = fig.add_axes([0,0,1,1])
ax.plot(years, df3)
plt.xticks(years)
plt.yticks(scale)
plt.ylabel("Total Number Of Officers", fontsize=18)
plt.xlabel("Years", fontsize=18)
plt.title("Married Military Officers In Each Year", fontsize=23);
```

For the second visualization, We have ploted the number of single military officers for each year, and we can see over time the graph have been steadily increasing, and the graph has reached its peak with 5323 single military officers in 2018. 

```python
df2 = df.groupby("Year").mean()
df3 = df2['Single']
years = [1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
scale = [2000,3000,4000,5000,6000]
fig=plt.figure(figsize=(13, 10))
ax = fig.add_axes([0,0,1,1])
ax.plot(years, df3)
plt.xticks(years)
plt.yticks(scale)
plt.ylabel("Total Number Of Officers", fontsize=18)
plt.xlabel("Years", fontsize=18)
plt.title("Single Military Officers In Each Year", fontsize=23);
```

For the third visualization, we have plotted the total number of officers in the military, similarly the first graph there was a large increase between 2000 and 2003 and after that point it is similar to the second graph where it increase steadily over time.

```python
yearlyTotal = df['Total']
plt.figure(figsize=(15,10))
plt.plot(range(1999,2019),yearlyTotal)
plt.title('Total Number of Officers in the Military',fontsize=23)
plt.xlabel('YEAR',fontsize=18)
plt.ylabel('Total number of officers',fontsize=18)
plt.xticks(np.arange(1999, 2019+1, 1.0))
plt.show();
```

For the last visualization, we have plotted the percentage of married and single officers in the military, while the total number of officers in the military are increasing, we can see the percentage of the married officers are decreasing, while percentage of single officers are moderately increasing.

```python
yearlySingle = df['Single']
yearlyMarried = df['Married']
PercSingle = []
PercMarried = []
for i in range(len(yearlySingle)):
    PercSingle.append(float(yearlySingle[i]/yearlyTotal[i]))
for i in range(len(yearlyMarried)):
    PercMarried.append(float(yearlyMarried[i]/yearlyTotal[i]))
plt.figure(figsize=(15,10))
plt.title("Percentage of Married and Single Officers in the Military",fontsize=23)
plt.plot(range(1999,2019),PercSingle, label = 'Percentage of Single')
plt.plot(range(1999,2019),PercMarried, label = 'Percentage of Married')
plt.xlabel('YEAR',fontsize=18)
plt.ylabel('Pecentage',fontsize=18)
plt.legend()
plt.xticks(np.arange(1999, 2019+1, 1.0))
plt.show();
```
